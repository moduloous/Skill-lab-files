```python
rho = 1.225
s = 10.0
Cl = 0.5

velocities = [10, 20, 30, 40, 50]

for v in velocities:
    lift = 0.5 * rho * v**2 * s * Cl
    print(f"Velocity: {v} m/s, Lift: {lift:.2f} N")
```

```python
density = [1.2, 1.0, 0.8]
height = [10, 20, 30]
gravity = 9.81

for rho, h in zip(density, height):
    pressure = rho * gravity * h
    print(f"Density: {rho} kg/m^3, Height: {h} m, Pressure: {pressure:.2f} Pa")
```

```python
# Define some variables
density = [1.2, 1.0, 0.8]  # Densities
velocities = [10, 20, 30] # Velocities
s_values = [10.0, 12.0]   # Wing areas
Cl_values = [0.5, 0.6]    # Lift coefficients

# Placeholder variables (assuming 'i' and 'g' are not standard physical properties in this context)
i_values = [1, 2]
g_values = [9.81, 9.78] # Example gravity values

# Using zip to pair some variables
for rho, v in zip(density, velocities):
    # Nested loop to consider different wing properties and placeholders
    for s, Cl, i, g in zip(s_values, Cl_values, i_values, g_values):
        # Example calculation (e.g., lift calculation combined with other variables)
        lift_or_other_calc = 0.5 * rho * v**2 * s * Cl * i * g # Example combining variables
        print(f"Density: {rho:.2f}, Velocity: {v}, Wing Area: {s}, Cl: {Cl}, i: {i}, g: {g:.2f}, Result: {lift_or_other_calc:.2f}")
```

```python
# Define constants
s = 10.0   # Wing area (m^2)
Cl = 0.5   # Lift coefficient

# Define variables
densities = [1.2, 1.0, 0.8]  # Densities
velocities = [10, 20, 30] # Velocities

# Nested loop structure
for i, rho in enumerate(densities):
    print(f"Outer loop - Index: {i}, Density: {rho}")
    # Inner loop over velocities (assuming this is the intended inner loop)
    for v in velocities:
        # Example calculation using constants and variables
        result = 0.5 * rho * v**2 * s * Cl
        print(f"  Inner loop - Velocity: {v}, Result: {result:.2f}")
```

```python
import numpy as np

# Create a matrix (2x3)
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])

# Create another matrix (3x2)
matrix_b = np.array([[7, 8],
                     [9, 10],
                     [11, 12]])

# Display the matrices
print("Matrix A:")
display(matrix_a)

print("\nMatrix B:")
display(matrix_b)

# Perform matrix multiplication (if dimensions are compatible)
if matrix_a.shape[1] == matrix_b.shape[0]:
    matrix_c = np.dot(matrix_a, matrix_b)
    print("\nMatrix A * Matrix B:")
    display(matrix_c)
else:
    print("\nMatrices cannot be multiplied due to incompatible dimensions.")
```
