```python
def calculate_acceleration(u, v, t):
  """Calculates acceleration using the formula (v - u) / t."""
  acceleration = (v - u) / t
  return acceleration

# Define different cases for initial velocity 'u', final velocity 'v', and time 't'
cases = [
    {"u": 10, "v": 50, "t": 5},
    {"u": 20, "v": 100, "t": 20},
    {"u": 0, "v": 30, "t": 10}
]

# Calculate acceleration for each case using a for loop
for case in cases:
  u = case["u"]
  v = case["v"]
  t = case["t"]
  acceleration = calculate_acceleration(u, v, t)
  print(f"Case: u={u}, v={v}, t={t}, Acceleration: {acceleration}")
```

```python
def calculate_aircraft_fuel(distance, speed, weight):
  """Calculates estimated fuel consumption based on distance, speed, and weight."""
  base_fuel_per_km = 0.05  # Base fuel consumption per kilometer (arbitrary unit)
  fuel_consumption = distance * base_fuel_per_km

  # Adjust fuel consumption based on speed
  if speed > 800:
    fuel_consumption *= 1.2  # Increase fuel for higher speeds
  elif speed < 500:
    fuel_consumption *= 0.9  # Decrease fuel for lower speeds

  # Adjust fuel consumption based on weight
  if weight > 50000:
    fuel_consumption *= 1.3  # Increase fuel for heavier aircraft
  elif weight < 30000:
    fuel_consumption *= 0.8  # Decrease fuel for lighter aircraft

  return fuel_consumption

# Example usage with different scenarios
scenario1_fuel = calculate_aircraft_fuel(distance=1000, speed=700, weight=40000)
print(f"Scenario 1 Fuel: {scenario1_fuel} units")

scenario2_fuel = calculate_aircraft_fuel(distance=500, speed=900, weight=60000)
print(f"Scenario 2 Fuel: {scenario2_fuel} units")

scenario3_fuel = calculate_aircraft_fuel(distance=2000, speed=400, weight=25000)
print(f"Scenario 3 Fuel: {scenario3_fuel} units")
```
