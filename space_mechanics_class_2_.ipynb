```python
import pandas as pd
import numpy as np

# Load your dataset from GitHub
air = pd.read_csv("https://raw.githubusercontent.com/cmvigneswaranaero-gif/AC_PERFROMANCE_WITH_ML/refs/heads/main/New_Atm_Para.csv")

air.info()  # Check data details

# Display first few rows for inspection
air.head()
```

```python
#Find the correlation between Altitude & pressure.
np.corrcoef(air.Altitude, air.Pressure)
```

```python
# -------------------------------------------------------------
# Find correlations between selected atmospheric parameters
# -------------------------------------------------------------

# 1. Altitude vs Temperature
corr_alt_temp = np.corrcoef(air.Altitude, air.Temperature)
print(f"Correlation between Altitude and Temperature:\n {corr_alt_temp}")

# 2. Pressure vs Density
corr_press_dens = np.corrcoef(air.Pressure, air.Density)[0,1]
# [0,1] extracts the correlation between the two variables.
print(f"Correlation between Pressure and Density: {corr_press_dens:.3f}")

# 3. Turbulence_Intensity vs Kinematic Viscosity
corr_mach_nu = np.corrcoef(air.Turbulence_Intensity, air.Kinematic_Viscosity)[0,1]
print(f"Correlation between Turbulence_Intensity and Kinematic Viscosity: {corr_mach_nu:.3f}")
```

```python
# -------------------------------------------------------------
# Optional: Display correlation matrix for all numerical features
# -------------------------------------------------------------
print("\nFull Correlation Matrix:")
print(air.corr().round(3))
# Computes the correlation matrix between all numerical columns in the DataFrame air.
# Rounds each correlation value to 3 decimal places for neat output.
```

```python
import pandas as pd
import statsmodels.formula.api as sm

# Load your dataset
air = pd.read_csv("https://raw.githubusercontent.com/cmvigneswaranaero-gif/AC_PERFROMANCE_WITH_ML/main/New_Atm_Para.csv")

# 1. Model 1: Altitude vs Density

model1 = sm.ols(formula='Density ~ Altitude', data=air)
fitted1 = model1.fit()
print("\nRegression Summary: Altitude vs Density")
print(fitted1.summary())
```

```python
import pandas as pd

# Create a new data point for prediction
new_data = pd.DataFrame({'Altitude': [9872]})

# Use your previously fitted model (fitted1)
predicted_density = fitted1.predict(new_data)

print("\nPredicted Density at 9872 m:")
print(predicted_density)
```
